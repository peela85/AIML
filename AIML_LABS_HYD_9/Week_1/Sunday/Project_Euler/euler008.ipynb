```python
import numpy as np
```

```python
grid = np.array([int(x) for x in data.split()])
```

```python
grid.shape
```

```python
grid = grid.reshape(20, 20)
```

```python
grid.shape
```

```python
grid[:,0:4]
```

```python
def hprod(row):
    return [np.prod(grid[row, col:col+4]) for col in range(0, 17)]
```

```python
def rowmax():
    for row in range(17):
        return max([max(hprod(row))])
```

```python
hprod(0)
```

```python
rowmax()
```

```python
def vprod(col):
    return [np.prod(grid[row:row+4, col]) for row in range(0, 17)]
```

```python
def colmax():
    for col in range(17):
        return max([max(vprod(col))])
```

```python
colmax()
```

```python
def rdprod(row, col):
    return np.prod([grid[x] for x in zip(list(range(row, min(row+4, grid.shape[0]))), list((range(col, col+4))))])
```

```python
row, col = 0, 0
rdprod(row, col)
    
```
